```python
import pandas as pd
```

```python
rootPath = "data/raw/movielens/"

movies = rootPath + "movies.csv"
links = rootPath + "links.csv"
ratings = rootPath + "ratings.csv"
tags = rootPath + "tags.csv"
```

```python
moviesDf = pd.read_csv(movies)
moviesDf
```

```python
linksDf = pd.read_csv(links)
linksDf[['imdbId']] = linksDf[['imdbId']].applymap(lambda x: 'https://www.imdb.com/title/tt' + str('{0:07}'.format(x)))
del linksDf['tmdbId']
linksDf
```

```python
ratingsDf = pd.read_csv(ratings)
ratingsDf
```

```python
tagsDf = pd.read_csv(tags)
tagsDf
```

```python
movies_links_df = pd.merge(moviesDf, linksDf, how='inner', on=['movieId'])
movies_ratings_df = pd.merge(movies_links_df, ratingsDf, how='inner', on=['movieId'])
movies_all_df = pd.merge(movies_ratings_df, tagsDf, how='inner', on=['movieId', 'userId'])
movies_all_df.columns = ['movielens_movieId', 'movielens_title', 'movielens_genres', 'imdb_link', 'movielens_userId', 'movielens_rating', 'movielens_timestamp_rating', 'movielens_user_review', 'movielens_timestamp_tag']
movies_all_df
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
def pprint(field, value):
    print(color.BOLD + str(field) + ": " + color.END + str(value) + "\n")

def movie_report(movielens_title):
    records = movies_all_df.loc[movies_all_df['movielens_title'] == movielens_title]
    pprint("Movie Title", movielens_title)
    pprint("IMDB", records['imdb_link'].iloc[0])
    pprint("Genres", records['movielens_genres'].iloc[0])
    pprint("Number of records found", records.shape[0])
    pprint("Average Rating", str(round(records['movielens_rating'].mean(), 2)) + " / 5.00")
    print(color.BOLD + "Reviews:" + color.END + "\n")
    temp = [print(text) for text in records['movielens_user_review']]
```

```python
movie_report('Avatar (2009)')
```

```python
movie_report('Avengers: Infinity War - Part I (2018)')
```

```python
movie_report('Jumanji: Welcome to the Jungle (2017)')
```

```python
movie_report('John Wick (2014)')
```
