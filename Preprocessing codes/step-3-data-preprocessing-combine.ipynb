```python
import pandas as pd
```

```python
movielensPreprocessedPath = 'data/preprocessed/movielens-only.csv'
imdbPreprocessedPath = 'data/preprocessed/imdb-only.csv'
outputPath = 'data/preprocessed/movielens-imdb-combined.csv'
```

```python
movielensDf = pd.read_csv(movielensPreprocessedPath)
movielensDf[['imdb_link']] = movielensDf[['imdb_link']].applymap(lambda x: x.strip())
movielensDf
```

```python
imdbPreprocessedDf = pd.read_csv(imdbPreprocessedPath)
imdbPreprocessedDf[['imdb_link']] = imdbPreprocessedDf[['imdb_link']].applymap(lambda x: x.strip())
imdbPreprocessedDf
```

```python
combined_df = pd.merge(movielensDf, imdbPreprocessedDf, how='inner', on=['imdb_link'])
combined_df
```

```python
print("No. of unique movie titles: " + str(combined_df['imdb_movie_title'].nunique()))
```

```python
print("Complete movie titles list:\n")
temp = [print(title) for title in sorted(combined_df['imdb_movie_title'].unique())]
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
    
def pprint(field, value):
    print(color.BOLD + str(field) + ": " + color.END + str(value) + "\n")
```

```python
def movie_report(imdb_title):
    records = combined_df.loc[combined_df['imdb_movie_title'] == imdb_title]
    pprint("IMDB Movie Title", imdb_title)
    pprint("Movielens Title", records['movielens_title'].iloc[0])
    pprint("Number of reviews found", records.shape[0])
    pprint("Year", int(records['imdb_title_year'].iloc[0]))
    pprint("IMDB", records['imdb_link'].iloc[0])
    pprint("Genres (IMDB)", records['imdb_genres'].iloc[0])
    pprint("Genres (Movielens)", records['movielens_genres'].iloc[0])
    pprint("Plot Keywords", records['imdb_plot_keywords'].iloc[0])
    pprint("Country", records['imdb_country'].iloc[0])
    pprint("Language", records['imdb_language'].iloc[0])
    pprint("Director Name (Fb Likes)", records['imdb_director_name'].iloc[0] + " (" + str(int(records['imdb_director_facebook_likes'].iloc[0])) + ")")
    pprint("Actor 1 (Fb Likes)", records['imdb_actor_1_name'].iloc[0] + " (" + str(int(records['imdb_actor_1_facebook_likes'].iloc[0])) + ")")
    pprint("Actor 2 (Fb Likes)", records['imdb_actor_2_name'].iloc[0] + " (" + str(int(records['imdb_actor_2_facebook_likes'].iloc[0])) + ")")
    pprint("Actor 3 (Fb Likes)", records['imdb_actor_3_name'].iloc[0] + " (" + str(int(records['imdb_actor_3_facebook_likes'].iloc[0])) + ")")
    pprint("Duration (in minutes)", int(records['imdb_duration'].iloc[0]))
    pprint("Cast total Facebook likes", int(records['imdb_cast_total_facebook_likes'].iloc[0]))
    pprint("Budget", records['imdb_budget'].iloc[0])
    pprint("Gross", records['imdb_gross'].iloc[0])
    pprint("Number of voted users", int(records['imdb_num_voted_users'].iloc[0]))
    pprint("Number of users for reviews", int(records['imdb_num_user_for_reviews'].iloc[0]))
    pprint("Number of critics for reviews", int(records['imdb_num_critic_for_reviews'].iloc[0]))
    pprint("IMDB Score", str(records['imdb_score'].iloc[0]) + " / 10.0")
    pprint("Movie Facebook likes", int(records['imdb_movie_facebook_likes'].iloc[0]))
    pprint("Average Movielens Rating", str(round(records['movielens_rating'].mean(), 2)) + " / 5.00")
    print(color.BOLD + "Movielens Reviews:" + color.END + "\n")
    temp = [print(text) for text in records['movielens_user_review']]
```

```python
movie_report('Avatar')
```

```python
movie_report('Titanic')
```

```python
combined_df.to_csv(outputPath, index=False)
```
